```python
#Question 1.
import numpy as np
import pandas as pd
import sklearn
dataset_train = pd.read_csv("titanic_train.csv")
dataset_test = pd.read_csv("titanic_test.csv")
d = dataset_train.append(dataset_test)
d.head()
```

```python
#Question 2.
#Remove missing values from Survived cloumns.
print("no.of missing values before ",d['Survived'].isnull().sum(axis=0))
d.dropna(subset=['Survived'],inplace=True)
print("no.of missing values after ",d['Survived'].isnull().sum(axis=0))
```

```python
#Question 3.
d['Has_Cabin'] = d['Cabin'].isnull()
```

```python
#Question 4.
import re
d['Title']=d['Name'].apply(lambda x:re.search('([A-Z][a-z]+\.)',x))
d['Title']=d['Title'].apply(lambda x:x.group())
d['Title']=d['Title'].replace(['Dr.','Rev.','Major.','Mlle.','Col.','Sir.','Jonkheer.','Capt.','Lady.',\
                               'Countess.','Ms.','Don.','Mme.'],'Special')
d['Title'].value_counts()
```

```python
#Question 5.
d.fillna(d.mean(), inplace=True)
d['Embarked'].fillna(d["Embarked"].mode(),inplace=True)
d['Cabin'].fillna(d["Cabin"].mode(),inplace=True)
```

```python
#Question 6.
d['Categorical_age'] = pd.qcut(d['Age'], 10,duplicates='drop')
d['Categorical_fare'] = pd.qcut(d['Fare'], 10)
```

```python
#Question 7.
d.drop(columns=['Cabin','Name','PassengerId','Ticket','Age','Fare','SibSp','Parch'])
```

```python
#Question 8.
pd.get_dummies(d.columns)
```

```python
#Question 9.
li = list(set(d.columns)-set(['Survived','Sex','PassengerId','Name','Ticket','Cabin',\
                              'Embarked','Title','Has_Cabin','Categorical_fare','Categorical_age']))
x = d[li]
y = d['Survived']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
#Question 10.
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
y_pred = regressor.predict(X_test)
pd.DataFrame({'Actual':y_test,"Predicted":y_pred})
```

```python
#Question 11.
sklearn.metrics.accuracy_score(y_test, y_pred.round(),normalize=True)
```
