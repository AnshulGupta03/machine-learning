```python
#Question1.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("weather1.csv")
```

```python
#Displaying histogram and PDF both
sns.distplot(df.Evaporation)
```

```python
#Displaying histogram only
sns.distplot(df.MaxTemp,kde=False,rug=True)
```

```python
#Displaying PDF only
sns.distplot(df.MaxTemp,hist=False,rug=True)
```

```python
#Displaying shaded PDF 
sns.kdeplot(df.WindGustSpeed,shade=True)
```

```python
#Displaying PDF with different binning 
sns.kdeplot(df.MinTemp,bw=3,label='bw=3')
sns.kdeplot(df.MaxTemp,bw=5,label='bw=5')
plt.legend()
```

```python
# Displaying PDF with different binning
sns.kdeplot(df.MinTemp,bw=6,label='bw=6')
sns.kdeplot(df.MaxTemp,bw=6,label='bw=6')
plt.legend()
```

```python
#Bivariate Analysis
df.plot(kind='Scatter',x='MinTemp',y='MaxTemp')
```

```python
#Bivariate Analysis
sns.jointplot(x='MinTemp',y='MaxTemp',data=df)
```

```python
#Hexbin plot
sns.jointplot(x=df.MinTemp,y=df.MaxTemp,kind='hex',color='white')
```

```python
#KernelDensityEstimation
sns.jointplot(x='Pressure9am',y='Cloud9am',data=df,kind='kde')
```

```python
#Pairplot
iris=sns.load_dataset('iris')
sns.pairplot(iris)
```

```python
#Strip plot
sns.stripplot(x='RainToday',y='Rainfall',data=df,jitter=True)
```

```python
#Swarmplot
sns.swarmplot(x='RainToday',y='Rainfall',data=df)
```

```python
#Box Plot
sns.boxplot(x='RainToday',y='Rainfall',data=df)
```

```python
#Bar plot
sns.barplot(x='Cloud9am',y='Rainfall',hue='RainTomorrow',data=df)
```
