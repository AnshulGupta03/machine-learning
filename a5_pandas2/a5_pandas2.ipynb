```python
#Question a.
import pandas as pd
file = pd.read_csv("weather1.csv")
print(file.isnull().sum())
```

```python
#Question b.
df1 = file.dropna(axis=1)
print(df1.isnull().sum())
```

```python
#question c.
df = pd.read_csv("weather1.csv")
print(df.isnull().sum())
print(50*"*","\n\nafter replacing missing values")
df.fillna(df.mean(),inplace = True)
df = df.apply(lambda x:x.fillna(x.value_counts().index[0]))
print(df.isnull().sum())
```
