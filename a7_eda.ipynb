```python
#Question 4.
#bivariate analysis
import pandas as pd
import seaborn as sns
df = pd.read_csv("house_data.csv")
sns.jointplot(x=df['bedrooms'],y=df['price'],data=df,kind='scatter')
'''
In the graph shown below we see that on increasing the numbers of bedrooms price will also increases.
and here we find an outlier also.
'''
```

```python
#bivariate analysis
sns.barplot(x="floors", y="price", data=df)
'''
In this graph we see that on increasing no. of floors ,price are also increasing then there is sudden decrease in price .
So, we can say that prices are not directly dapendent on floors.
'''
```

```python
#bivariate analysis
sns.barplot(x="floors", y="price", hue='view',data=df)
```

```python
#bivariate analysis
sns.barplot(x="view", y="price", data=df)
'''
The below shown visualization is a barplot in which prices are increasing gradually on increasing view.
'''
```
